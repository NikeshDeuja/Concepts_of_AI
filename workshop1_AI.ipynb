```python
def border_matrix(n):
    m = np.zeros((n,n), dtype=int)
    m[0,:] = 1
    m[-1,:] = 1
    m[:,0] = 1
    m[:,-1] = 1
    return m

border_5 = border_matrix(5)
print("\n5x5 border matrix:\n", border_5)
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
I = np.eye(2, dtype=float)

A_inv = np.linalg.inv(A)
prod_A_Ainv = A.dot(A_inv)
print("A * A_inv:\n", prod_A_Ainv)
```

```python
AB = A.dot(B)
BA = B.dot(A)
print("\nA*B:\n", AB)
print("\nB*A:\n", BA)
print("\nAre AB and BA equal?", np.allclose(AB, BA))
```

```python
lhs = (A.dot(B)).T
rhs = B.T.dot(A.T)
print("\n(AB)^T:\n", lhs)
print("\nB^T A^T:\n", rhs)
print("\n(AB)^T == B^T A^T ?", np.allclose(lhs, rhs))
coeff = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]], dtype=float)
rhs_vec = np.array([-1, -3, 9], dtype=float)

solution = np.linalg.solve(coeff, rhs_vec)
print("\nSolution (using np.linalg.solve):", solution)

coeff_inv = np.linalg.inv(coeff)
solution_via_inv = coeff_inv.dot(rhs_vec)
print("Solution (via inverse):", solution_via_inv)

print("Solutions equal?", np.allclose(solution, solution_via_inv))
```

```python
import random
import time

N = 1_000_000
M = 1000
list1 = [random.random() for _ in range(N)]
list2 = [random.random() for _ in range(N)]

# Python lists addition (element-wise)
t0 = time.time()
list_add = [list1[i] + list2[i] for i in range(N)]
t1 = time.time()
print("Python lists element-wise addition time:", t1 - t0, "seconds")

# NumPy arrays addition
a1 = np.array(list1)
a2 = np.array(list2)
t0 = time.time()
np_add = a1 + a2
t1 = time.time()
print("NumPy arrays addition time:", t1 - t0, "seconds")
```

```python
# 2. Element-wise multiplication
t0 = time.time()
list_mul = [list1[i] * list2[i] for i in range(N)]
t1 = time.time()
print("Python lists element-wise multiplication time:", t1 - t0, "seconds")

t0 = time.time()
np_mul = a1 * a2
t1 = time.time()
print("NumPy arrays multiplication time:", t1 - t0, "seconds")
```

```python
# 3. Dot product
t0 = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(N))
t1 = time.time()
print("Python lists dot product time:", t1 - t0, "seconds")

t0 = time.time()
dot_np = a1.dot(a2)
t1 = time.time()
print("NumPy dot product time:", t1 - t0, "seconds")
```

```python
# 4. Matrix multiplication 1000x1000
A_np = np.random.random((M,M))
B_np = np.random.random((M,M))

t0 = time.time()
C_np = A_np.dot(B_np)
t1 = time.time()
print(f"NumPy matrix multiplication {M}x{M} time:", t1 - t0, "seconds")

M_small = 200
A_list = [[random.random() for _ in range(M_small)] for _ in range(M_small)]
B_list = [[random.random() for _ in range(M_small)] for _ in range(M_small)]

t0 = time.time()
C_list = [[0.0]*M_small for _ in range(M_small)]
for i in range(M_small):
    for k in range(M_small):
        aik = A_list[i][k]
        rowC = C_list[i]
        rowB = B_list[k]
        for j in range(M_small):
            rowC[j] += aik * rowB[j]
t1 = time.time()
print(f"Pure-Python list matrix multiplication {M_small}x{M_small} time:", t1 - t0, "seconds")
print("Note: pure-Python 1000x1000 would take a very long time; NumPy is the practical approach.")
```
